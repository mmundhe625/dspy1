1. What is need for Hypothesis Testing ?

ANS:-
1.To make decisions using data
  Instead of guessing, hypothesis testing helps us take decisions based on scientific evidence.
2.To test assumptions (hypotheses)
  It checks whether a statement (like average, proportion, etc.) is true or not.
3.To avoid random conclusions
  It helps us understand whether results are due to actual effects or just chance.
4.To compare two situations
  For example:
  * Old method vs new method
  * Before vs after experiment
5.To support research and experiments
  Used in fields like:
       * Science
       * Engineering
       * Medicine
       * Business
6.To control errors in decision-making
  It reduces chances of making wrong conclusions using concepts like:
  * Level of significance (α)
  * Type I and Type II errors


  Example: Test if Average Age = 19

```python
# Import libraries
import numpy as np
from scipy import stats

# Sample data (ages of students)
ages = [18, 19, 20, 21, 19, 18, 20, 22, 19, 21]

# Hypothesized population mean
mu = 19

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(ages, mu)

# Print results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision making
alpha = 0.05  # significance level

if p_value < alpha:
    print("Reject Null Hypothesis (H0)")
else:
    print("Fail to Reject Null Hypothesis (H0)")
```

3. What is a student t distribution?

ANS:-

The Student's t-distribution is a probability distribution used in statistics when:

* the sample size is small (n < 30), and
* the population standard deviation (σ) is unknown.

It was introduced by William Sealy Gosset under the pen name “Student.”

Formula

The t-statistic is given by:

$$
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
$$
   







Where:


ˉ
x
ˉ
 = sample mean


μ = population mean


s = sample standard deviation

n = sample size



Key Characteristics:

* Bell-shaped and symmetric (like normal distribution)
* Has heavier tails → more spread out than normal distribution
* Depends on degrees of freedom (df = n - 1)
* As sample size increases → it approaches normal distribution

When to Use:

* Small sample size
* Population standard deviation is unknown
* Data is approximately normally distributed

4. What is level of significance ? What role t plays in hypothesis Testing

ANS:-


Definition:

The level of significance (α) is the maximum probability of rejecting the null hypothesis (H₀) when it is actually true.
It is the risk of making a wrong decision (Type I error).
Common Values:


  * α = 0.05 (5%)
  * α = 0.01 (1%)

Role of t in Testing:
   
   1.Measures difference

       * Shows how far sample mean is from population mean
   2.Decision Making
   
       * Compare calculated t with critical t-value
       * OR use p-value
   3.Rule:

       * If |t| > critical value → Reject H₀
       * If |t| ≤ critical value → Do not reject H₀
  4.Used in t-tests

       * One-sample t-test
       * Two-sample t-test
       * Paired t-test

5. Design a hypothesis testing for identifying average age of students in TE IT class is 19 or not.

      * Null Hypothesis (H₀): μ = 19
      * Alternative Hypothesis (H₁): μ ≠ 19
Step 2: Level of Significance

      * Take α = 0.05 (5%)
Step 3: Choose Test Statistic

$$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$$

Step 4: Determine Critical Value
      
      * Degrees of freedom: df = n - 1

Step 5: Decision Rule
      
      * If |t| > t₍critical₎ → Reject H₀
      * If |t| ≤ t₍critical₎ → Do not reject H₀      

Step 6: Conclusion

```python
import numpy as np
from scipy import stats

# Sample data
data = np.array([18, 19, 20, 19, 18, 21, 19, 20, 18, 19])

# Hypothesized mean
mu = 19

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(data, mu)

print("t-statistic:", t_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
